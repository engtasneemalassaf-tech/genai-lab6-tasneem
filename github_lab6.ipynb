```python
from getpass import getpass
GITHUB_TOKEN = getpass("Paste your GitHub Personal Access Token (it will be hidden): ")
print("Token stored in memory (not printed).")
```

```python
from openai import OpenAI
client = OpenAI(
    base_url="https://models.github.ai/inference/v1",
    api_key=GITHUB_TOKEN
)
print("Client created successfully.")
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "Hello from Google Colab! Write one fun fact about AI."}
    ]
)
print("Model response:\n")
print(response.choices[0].message.content)
```

```python
paragraph = """Generative AI represents a significant advancement in artificial intelligence, enabling machines to create content that mimics human creativity. This technology utilizes complex algorithms and vast datasets to produce text, images, music, and more, often indistinguishable from human-generated works. Its applications span various fields, including entertainment, marketing, and education, where it can enhance productivity and innovation. However, the rise of generative AI also raises ethical concerns regarding originality, copyright, and the potential for misuse, necessitating ongoing discussions about its implications in society."""
prompt = f"Summarize the following paragraph into 3 concise bullet points:\n\n{paragraph}"

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role":"user","content":prompt}]
)

print("Summary:")
print(response.choices[0].message.content)
```
